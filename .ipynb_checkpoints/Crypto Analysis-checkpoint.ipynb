```python
import pandas_datareader as web
import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

start = dt.datetime(2016, 1, 1)
end = dt.datetime.now()

cro = web.DataReader('CRO-USD', 'yahoo', start, end)
cro.head(50)
```
